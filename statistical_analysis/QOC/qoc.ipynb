```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
qoc_data = pd.read_csv('qoc_data.csv')
```

```python
qoc_data.head()
```

```python
qoc_data = qoc_data.rename(columns={'Soft%':'Soft', 'Med%':'Med', 'Hard%':'Hard'})
```

```python
qoc_data.head()
```

```python
qoc_data['Soft'] = qoc_data['Soft'].str.replace('%', '')
qoc_data['Med'] = qoc_data['Med'].str.replace('%', '')
qoc_data['Hard'] = qoc_data['Hard'].str.replace('%', '')
```

```python
qoc_data.head()
```

```python
soft = qoc_data.Soft
med = qoc_data.Med
hard = qoc_data.Hard
year = qoc_data.Season
```

```python
df = pd.DataFrame()
```

```python
df.insert(0, 'Soft', soft)
df.insert(1, 'Med', med)
df.insert(2, 'Hard', hard)
```

```python
df.insert(0, 'Year', year)
```

```python
df
```

```python
df.set_index('Year')
```

```python
df = df.reset_index(drop=True)
```

```python
df.set_index('Year', inplace=True)
```

```python
df
```

```python
qoc_data = df
```

```python
qoc_data.to_csv('qoc_data.csv')
```
