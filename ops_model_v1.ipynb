```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

```python
baseball_data = pd.read_csv('baseball_data.csv')
```

```python
baseball_data.head(5)
```

```python
baseball_data = baseball_data.drop(columns=['Season', 'Team'])
```

```python
baseball_data = baseball_data.rename(columns={'BB%':'BB', 'K%':'K', 'wRC+':'WRCplus'})
```

```python
baseball_data.head(5)
```

```python
baseball_data['BB'] = baseball_data['BB'].str.replace('%', '')
```

```python
baseball_data['K'] = baseball_data['K'].str.replace('%', '')
```

```python
baseball_data.head(5)
```

```python
# Prediction target for model

y = baseball_data.OPS
```

```python
baseball_data.columns
```

```python
features = ['G', 'PA', 'HR', 'R', 'RBI', 'SB', 'BB', 'K', 'ISO', 'BABIP', 'AVG', 'OBP', 'SLG', 'OPS', 'wOBA', 'WRCplus', 'WAR']
```

```python
# Features for model

X = baseball_data[features]
```

```python
X.describe
```

```python
y.describe
```

```python
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=0)
```

```python
ops_model = DecisionTreeRegressor(random_state=0)
```

```python
ops_model.fit(train_X, train_y)
```

```python
val_predictions = ops_model.predict(val_X)
```

```python
print(mean_absolute_error(val_y, val_predictions))
```

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```

```python
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```
